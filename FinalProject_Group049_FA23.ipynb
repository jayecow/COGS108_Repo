We hypothesize that the number of years in college, FT% (free throw percentage), and 3PT% (three-pointer percentage) are the most important factors to consider when predicting which players will get drafted into the NBA. Logically, it makes sense that players who have the most experience (more years in college) are more likely to perform well in basketball, and thus have better chances of getting drafted. ## Set Up

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_fscore_support, f1_score, accuracy_score
```

## College Basketball 2009-2021 + NBA Advanced Stats

```python
# Import the raw dataset
college_bb_raw = pd.read_csv('https://raw.githubusercontent.com/rhlin2001/COGS108_Repo/main/CollegeBasketballPlayers2009-2021.csv', dtype = {'num' : 'object'})
college_bb_raw.head()
```

```python
# Take subset of relevant columns and rename them
college_bb = college_bb_raw[['player_name', 'team', 'yr', 'FT_per', 'twoP_per', 'TP_per', 'pts', 'treb', 'ast']]
college_bb = college_bb.rename(columns={'player_name': 'player', 'yr': 'year', 'FT_per': 'ft%', 'twoP_per': '2pt%', 'TP_per': '3pt%', 'pts': 'ppg', 'treb': 'rpg', 'ast': 'apg'})
college_bb.head()
```

```python
# Observe some weird values for year column
college_bb['year'].value_counts()
```

```python
# Drop rows with weird year values
college_bb = college_bb.loc[college_bb['year'].isin(['Fr', 'So', 'Jr', 'Sr'])]
years_in_college = college_bb.drop_duplicates(subset=['player', 'team', 'year']).groupby(['player', 'team']).count().reset_index()['year']

# Group by player and team
aggregate_dict = {'ft%': 'mean', '2pt%': 'mean', '3pt%': 'mean', 'ppg': 'sum', 'rpg': 'sum', 'apg': 'sum'}
college_bb = college_bb.groupby(['player', 'team']).aggregate(aggregate_dict).reset_index()
college_bb = college_bb.assign(years = years_in_college)
college_bb.head()
```

```python
# No missing values
college_bb.isna().sum()
```

## NBA Draft Basketball Player Data 1989-2021

```python
# Import the raw dataset
nba_draft_raw = pd.read_csv('https://raw.githubusercontent.com/rhlin2001/COGS108_Repo/main/nbaplayersdraft.csv')
nba_draft_raw.head()
```

```python
# Take subset of relevant columns and assign new drafted column
nba_draft = nba_draft_raw[['player']]
nba_draft = nba_draft.assign(drafted = True)
nba_draft.head()
```

```python
# No missing values
nba_draft.isna().sum()
```

## Merge

```python
# Merge on player and populate missing values
college_draft = college_bb.merge(nba_draft, how='left', on='player')
college_draft['drafted'].fillna(False, inplace=True)
college_draft.head()
```

```python
# No missing values
college_draft.isna().sum()
```

## Predicting the NBA Draft Using SVM

Since SVM is sensitive to outliers, we begin by removing the outliers from our quantitative columns using the z-score method with a z-threshold of 3.

```python
# Remove outliers
college_draft_quant = college_draft[['ft%', '2pt%', '3pt%', 'ppg', 'rpg', 'apg', 'years']]
z_scores = np.abs((college_draft_quant - college_draft_quant.mean()) / college_draft_quant.std())
no_outliers = college_draft_quant[(z_scores < 3).all(axis=1)]
college_draft_adjust = college_draft.loc[no_outliers.index]
college_draft_adjust.head()
```

Since the nature of our data has some overlap between drafted and undrafted players, we use a Radial Basis Function (RBF) kernel. To address the significant imbalance between the proportions of drafted versus undrafted players, we utilize a balanced class weighting.

```python
# Build our SVM model
def train_SVM(X, y, kernel='rbf', class_weight='balanced'):
    clf = SVC(kernel=kernel)
    clf.fit(X, y)
    return clf
```

We use a 80:20 split for the training and testing sets respectively and select the drafted column as our labels.

```python
# Split training and testing data
num_training = int(len(college_draft_adjust) * 0.8)
num_testing = len(college_draft_adjust) - num_training

CD_Y = college_draft_adjust['drafted'].to_numpy()
CD_train_Y = CD_Y[:num_training]
CD_test_Y = CD_Y[num_training:]
```

Now we use all of our quantitative features as predictors to observe their combined performance.

```python
# SVM using all features
CD_all_X = college_draft_adjust[['ft%', '2pt%', '3pt%', 'ppg', 'rpg', 'apg', 'years']]
CD_all_train_X = CD_all_X[:num_training]
CD_all_test_X = CD_all_X[num_training:]

CD_all_clf = train_SVM(CD_all_train_X, CD_train_Y)
CD_all_pred_train_Y = CD_all_clf.predict(CD_all_train_X)
CD_all_pred_test_Y = CD_all_clf.predict(CD_all_test_X)

CD_all_class_report_train = classification_report(CD_train_Y, CD_all_pred_train_Y)
CD_all_conf_mat_train = confusion_matrix(CD_train_Y, CD_all_pred_train_Y)
print('Training Set Performance:')
print(CD_all_class_report_train)
print(CD_all_conf_mat_train)

CD_all_class_report_test = classification_report(CD_test_Y, CD_all_pred_test_Y)
CD_all_conf_mat_test = confusion_matrix(CD_test_Y, CD_all_pred_test_Y)
print('Testing Set Performance:')
print(CD_all_class_report_test)
print(CD_all_conf_mat_test)
```

Unfortunately, for both the training and testing sets, our model has 100% accuracy for classifying undrafted players, but 0% accuracy for classifying drafted players. We see this in our confusion matrix where we get 0 false positives, meaning that 0 undrafted players were incorrectly predicted as drafted. However, we also see 0 true positives, meaning that 0 drafted players were correctly predicted as drafted.

Next we use our measures of volume as predictors to observe their performance.

```python
# SVM using measures of volume
CD_vol_X = college_draft_adjust[['ppg', 'rpg', 'apg']]
CD_vol_train_X = CD_vol_X[:num_training]
CD_vol_test_X = CD_vol_X[num_training:]

CD_vol_clf = train_SVM(CD_vol_train_X, CD_train_Y)
CD_vol_pred_train_Y = CD_vol_clf.predict(CD_vol_train_X)
CD_vol_pred_test_Y = CD_vol_clf.predict(CD_vol_test_X)

CD_vol_class_report_train = classification_report(CD_train_Y, CD_vol_pred_train_Y)
CD_vol_conf_mat_train = confusion_matrix(CD_train_Y, CD_vol_pred_train_Y)
print('Training Set Performance:')
print(CD_vol_class_report_train)
print(CD_vol_conf_mat_train)

CD_vol_class_report_test = classification_report(CD_test_Y, CD_vol_pred_test_Y)
CD_vol_conf_mat_test = confusion_matrix(CD_test_Y, CD_vol_pred_test_Y)
print('Testing Set Performance:')
print(CD_vol_class_report_test)
print(CD_vol_conf_mat_test)
```

Similarly, for both the training and testing sets, our model has 100% accuracy for classifying undrafted players, but 0% accuracy for classifying drafted players.

Finally, we use our measures of efficiency as predictors to observe their performance.

```python
# SVM using measures of efficiency
CD_eff_X = college_draft_adjust[['ft%', '2pt%', '3pt%']]
CD_eff_train_X = CD_eff_X[:num_training]
CD_eff_test_X = CD_eff_X[num_training:]

CD_eff_clf = train_SVM(CD_eff_train_X, CD_train_Y)
CD_eff_pred_train_Y = CD_eff_clf.predict(CD_eff_train_X)
CD_eff_pred_test_Y = CD_eff_clf.predict(CD_eff_test_X)

CD_eff_class_report_train = classification_report(CD_train_Y, CD_eff_pred_train_Y)
CD_eff_conf_mat_train = confusion_matrix(CD_train_Y, CD_eff_pred_train_Y)
print('Training Set Performance:')
print(CD_eff_class_report_train)
print(CD_eff_conf_mat_train)

CD_eff_class_report_test = classification_report(CD_test_Y, CD_eff_pred_test_Y)
CD_eff_conf_mat_test = confusion_matrix(CD_test_Y, CD_eff_pred_test_Y)
print('Testing Set Performance:')
print(CD_eff_class_report_test)
print(CD_eff_conf_mat_test)
```

Again, for both the training and testing sets, our model has 100% accuracy for classifying undrafted players, but 0% accuracy for classifying drafted players.

To summarize, despite our efforts to use class weighting and different combinations of predictors, we are not able to build a model that can accurately predict drafted players. This is likely due to the fact that there is considerable overlap in our data between drafted and undrafted players which is an indication that our chosen features may not be the best predictors for the NBA Draft. Perhaps in future analyses, we may try to use other features to help predict the NBA Draft.

## Predicting the NBA Draft Using KNN

We will now try using another method, KNN, to build a model to predict drafted players to see if we can get any accurate results for both drafted and undrafted players.

KNN is less tolerant of outliers than SVM, so we will continue to use the adjusted dataset that removes outliers for this analysis. We will also create two subsets of data, one that includes the volume metrics of each player and the another for efficiency metrics, as well as their draft status.

```python
vol_college_draft_adjust = college_draft_adjust[['player','ppg', 'rpg', 'apg', 'drafted']]
eff_college_draft_adjust = college_draft_adjust[['player','ft%', '2pt%', '3pt%', 'drafted']]
```

We are further splitting the dataset here into two more sets, one that contains the features we will use to classify each data point (PPG, RPG, and APG), and another dataset with the players' draft status.

```python
vol_X = vol_college_draft_adjust.iloc[:,1:4]
vol_X = vol_X.values

vol_Y = vol_college_draft_adjust.iloc[:, -1]
vol_Y = vol_Y.values
```

Here we have defined our training and testing data. The training data will be used to create a model that uses our features we specified previously in order to predict each players' draft status. The trained model will be tried on the test data to check for precision in its predictions. We've continued to use an 80:20 split in the training and testing datasets.

```python
vol_train_X, vol_test_X, vol_train_Y, vol_test_Y = train_test_split(vol_X, vol_Y, test_size=0.2, random_state=42)
```

We are not working with categorical features, so we have chosen to use the euclidean distance (straight line between points) to classify them. And we will train our classifier here to predict drafted/undrafted players. We've chosen k by doing square root of the dataset size, divided in half, in order to avoid overfitting and underfitting the model.

```python
knn = KNeighborsClassifier(n_neighbors=167, p=2, metric='euclidean')
knn.fit(vol_train_X, vol_train_Y)
```

Using the trained classifier, we will predict whether players in the training and testing data sets were drafted based on if their volume metrics are more similar to drafted or undrafted players in the dataset. # Discussion and Conclusion

From our analysis, we were able to draw some conclusions about the statistical differences between players that get drafted and those who do not. Unsurprisingly, as our hypothesis suggested, players who were drafted were more efficient according to statistics we were tracking (FT%, 2pt%, and 3pt%). The low-end of drafted players' stats were often the median of undrafted players' stats. However, we didn't expect that volume stats (PPG, RPG, and APG) would be so drastically different between the two sets of players. In fact, the gulf between volume statistics was far wider than the gap between efficiency metrics for the two groups. This largely led to us rejecting our hypothesis; efficiency statistics were in fact not better predictors of draft status than volume statistics were.

This leads us to the conclusion that, while drafted players outperform their undrafted counterparts in all statistics, volume statistics are better indicators for the likelihood of whether or not someone will be drafted. Undrafted players averaged lower numbers across all six statistical categories we looked at but the gap was significantly larger in volume stats.


Despite these observations, we were also unable to create a model that convincingly determines whether or not a player will be drafted based on their statistics. This might be due to the large overlap between drafted and undrafted players. Additionally, there are some improvements to the study listed below that might have resulted in some more concrete takeaways.

If another group were to pick up where we left off, there are some improvements and suggestions we have for them. First of all, the difference in volume statistics might be attributed to the playing time of the players. Players with more playing time have more of an opportunity to tally statistics like points, rebounds, and assists. As such, a good approach might have been to average out these statistics per minute or to use the widely-accepted per-36 minute statistic.

Secondly, it would be more illuminating to take into consideration the physical aspects of players. Based on anecdotal observations, basketball players are drafted for "intangible" characteristics that are influenced by their physical attributes. Taller, faster players might be given more of a chance even if their statistics might not stack up to their competition in hopes that such players can become NBA-caliber.

Thirdly, it might have been interesting to use advanced statistics like true shooting (TS%) and effective field goal percentage (EFG%) in conjunction with the traditional efficiency stats we used. These additional stats would provide more insight as well as create a more standardized model to compare players on instead of having three disjoint categories. These statistics serve as an average of all types of shooting so the more-traditional statistics might not even be necessary.

Finally, we would suggest having some sort of positional analysis. Our approach didn't take different positions into account. This is important because the statistical expectations are different for a center vs a forward vs a guard. For instance, centers would be expected to have higher RPG while guards might be expected to have higher APG. We'd guess that the statistical differences between drafted and undrafted players would have less variance if compared by position instead of between all players.

Our project's societal impact would be most apparent in younger basketball players and their coaches. Players who are interested in becoming NBA players can prioritize maximizing their statistical impact in the fields we looked at to increase their likelihood of making the league. Likewise, youth coaches can apply our findings to better craft and organize their training methodologies to address these specific statistical categories.
